```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.csv')
```

```python
for c in df['company'].unique():
    print('Company:',c)
    tempdf = df[df['company'] == c].loc[:,['hour','max_high']]
    print(tempdf.set_index('hour').plot(kind='box'))
    plt.show()

```
