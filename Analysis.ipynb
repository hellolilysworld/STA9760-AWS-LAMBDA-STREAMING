# Analysis of Hourly highs for a basket of companies as of May 14, 2020

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.csv')
```

## Distribution of hourly high prices

```python
for c in df['company'].unique():
    print('Company:',c)
    tempdf = df[df['company'] == c].loc[:,['hour','max_high']]
    print(tempdf.set_index('hour').plot(kind='box'))
    plt.show()

```

## Trend of hourly high prices 

```python
for c in df['company'].unique():
    print('Company:',c)
    tempdf = df[df['company'] == c].loc[:,['hour','max_high']]
    print(tempdf.set_index('hour').plot(kind='line'))
    plt.show()

```
